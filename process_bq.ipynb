```python
import json

# Read the BQ results file
with open('C:/Users/jquirno/Downloads/bq-results-20260417-143838-1776436845982.json', 'r', encoding='utf-8') as f:
    raw = json.load(f)

# Handle both array and object forms
if isinstance(raw, list):
    data_json_str = raw[0]['data_json']
else:
    data_json_str = raw['data_json']

# Parse the inner JSON
inner = json.loads(data_json_str)

print('Keys:', list(inner.keys()))
print('ms rows:', len(inner.get('ms', [])))
print('landing rows:', len(inner.get('landing', [])))
print('container rows:', len(inner.get('container', [])))
if inner.get('ms'):
    print('First ms row:', json.dumps(inner['ms'][0]))
if inner.get('landing'):
    print('First landing row:', json.dumps(inner['landing'][0]))
if inner.get('container'):
    print('First container row:', json.dumps(inner['container'][0]))
```

```python
# Read current data.json
with open('C:/Users/jquirno/proyecto-claude/campaign_tracker/docs/data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Build historical entry
historical_entry = {
    'name': 'DOUBLE DATES 04/04',
    'start': '2026-03-23',
    'end': '2026-04-07',
    'ms': inner.get('ms', []),
    'landing': inner.get('landing', []),
    'container': inner.get('container', [])
}

# Push into historical array
data['historical'].append(historical_entry)

# Write updated data.json
with open('C:/Users/jquirno/proyecto-claude/campaign_tracker/docs/data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print('Done! historical entries:', len(data['historical']))
print('ms rows written:', len(historical_entry['ms']))
print('landing rows written:', len(historical_entry['landing']))
print('container rows written:', len(historical_entry['container']))
```
